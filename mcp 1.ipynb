### **2- Arreglo de los valores de ROI**

Teniendo en cuenta que ROI = Revenue / Spend, para aquellas filas que se pueda calcular lo calcularemos. Notese que no coincide el cálculo por lo que para asegurar el tiro recalcularemos la columna


```python
# Función para calcular o actualizar el ROI basado en Revenue y Spend
def calculate_roi(row):
    # Si Revenue y Spend no son nulos, calcular ROI como (Revenue / Spend) * 100
    if pd.notnull(row['Revenue']) and pd.notnull(row['Spend']):
        return (row['Revenue'] / row['Spend']) * 100
    # Si ROI ya tiene un valor, se multiplica por 100 para expresarlo en porcentaje
    else:
        return row['ROI'] * 100  # Suponiendo que el ROI ya estaba dividido

# Función para calcular o actualizar Spend basado en Revenue y ROI
def calculate_spend(row):
    # Si Revenue y ROI no son nulos, y Spend es nulo, calcular Spend como Revenue / (ROI / 100)
    if pd.notnull(row['Revenue']) and pd.notnull(row['ROI']) and pd.isnull(row['Spend']) and row['ROI'] != 0:
        return row['Revenue'] / (row['ROI'] / 100)
    # Si Spend ya tiene un valor, devolverlo sin cambios
    else:
        return row['Spend']

# Función para calcular o actualizar Revenue basado en Spend y ROI
def calculate_revenue(row):
    # Si Spend y ROI no son nulos, y Revenue es nulo, calcular Revenue como Spend * (ROI / 100)
    if pd.notnull(row['Spend']) and pd.notnull(row['ROI']) and pd.isnull(row['Revenue']) and row['ROI'] != 0:
        return row['Spend'] * (row['ROI'] / 100)
    # Si Revenue ya tiene un valor, devolverlo sin cambios
    else:
        return row['Revenue']

# Aplicar las funciones a las columnas del DataFrame
df['ROI'] = df.apply(calculate_roi, axis=1)
df['Spend'] = df.apply(calculate_spend, axis=1)
df['Revenue'] = df.apply(calculate_revenue, axis=1)
```

### **3- Arreglo de Impressions, Clicks, Conversions, CR y CTR**


Teniendo en cuenta las relaciones entre estas columnas vamos a intentar bajar el porcentaje de nulos 



```python
# Realizamos una función para calcular el CTR. Si Clicks e Impressions no son nulos, calculamos el CTR como Clicks/Impressions*100. Si Clicks es mayor que Impressions, intercambiamos los valores de Clicks e Impressions. 
# Si Clicks o Impressions son nulos, devolvemos el valor de CTR multiplicado por 100 para dejarlo por defecto en porcentaje.
def calculate_ctr(row):
    if pd.notnull(row['Impressions']) and pd.notnull(row['Clicks']):
        if row['Clicks'] >= row['Impressions']:
            # Intercambiar los valores de Clicks e Impressions
            row['Clicks'], row['Impressions'] = row['Impressions'], row['Clicks']
        return (row['Clicks'] / row['Impressions']) * 100
    else:
        return row['CTR'] * 100
# Realizamos una función para calcular el CR. Si Conversions y Clicks no son nulos, calculamos el CR como Conversions/Clicks*100. Si Conversions es mayor que Clicks, intercambiamos los valores de Conversions y Clicks.
# Si Conversions o Clicks son nulos, devolvemos el valor de CR multiplicado por 100 para dejarlo por defecto en porcentaje.
def calculate_cr(row):
    if pd.notnull(row['Conversions']) and pd.notnull(row['Clicks']):
        if row['Conversions'] >= row['Clicks']:
            # Intercambiar los valores de Conversions y Clicks
            row['Conversions'], row['Clicks'] = row['Clicks'], row['Conversions']
        return (row['Conversions'] / row['Clicks']) * 100
    else:
        return row['CR'] * 100

# Función para calcular y corregir valores nulos en la columna 'Conversions'
def calculate_conversions(row):
    # Caso 1: Calcular Conversions si es nulo, pero Clicks y CR tienen valores
    # Fórmula utilizada: Conversions = (CR / 100) * Clicks
    # Esto asegura que Conversions se calcula de acuerdo al valor de CR (Conversion Rate)
    if pd.isnull(row['Conversions']) and pd.notnull(row['Clicks']) and pd.notnull(row['CR']):
        return (row['CR'] / 100) * row['Clicks']
    
    # Caso por defecto: Devolver el valor existente en Conversions
    # Si Conversions ya tiene un valor, se devuelve sin cambios
    else:
        return row['Conversions']


# Función para calcular y corregir valores nulos en la columna 'Impressions'
def calculate_impressions(row):
    # Caso 1: Calcular Impressions si es nulo, pero Clicks y CTR tienen valores
    # Fórmula utilizada: Impressions = Clicks / (CTR / 100)
    # Esto asegura que Impressions se calcula de acuerdo al valor de CTR
    if pd.isnull(row['Impressions']) and pd.notnull(row['Clicks']) and pd.notnull(row['CTR']):
        return row['Clicks'] / (row['CTR'] / 100)
    
    # Caso por defecto: Devolver el valor existente en Impressions
    # Si Impressions ya tiene un valor, se devuelve sin cambios
    else:
        return row['Impressions']

    
# Función para calcular y corregir valores nulos en la columna 'Clicks'
def calculate_clicks(row):
    # Caso 1: Calcular Clicks usando CTR e Impressions si CR o Conversions son nulos
    if pd.isnull(row['Clicks']) and pd.notnull(row['CTR']) and pd.notnull(row['Impressions']) and (pd.isnull(row['CR']) or pd.isnull(row['Conversions'])):
        return (row['CTR'] / 100) * row['Impressions']
    
    # Caso 2: Calcular Clicks usando CR y Conversions si CTR o Impressions son nulos
    elif pd.isnull(row['Clicks']) and pd.notnull(row['CR']) and pd.notnull(row['Conversions']) and (pd.isnull(row['CTR']) or pd.isnull(row['Impressions'])):
        return row['Conversions'] / (row['CR'] / 100)
    
    # Caso por defecto: Devolver el valor existente en Clicks
    else:
        return row['Clicks']


df['CTR'] = df.apply(calculate_ctr, axis=1)

df['CR'] = df.apply(calculate_cr, axis=1)

df['Conversions'] = df.apply(calculate_conversions, axis=1)

df['Impressions'] = df.apply(calculate_impressions, axis=1)

df['Clicks'] = df.apply(calculate_clicks, axis=1)
```

### **5.1 - IMPUTACIÓN DE LAS VARIABLES NUMÉRICAS CON LA MEDIANA PARA CADA SUBCONJUNTO DEL DATASET**

Imputamos con la mediana las variables numéricas para cada combinación de datos de Region, Product_Category, Channel y Market_Segment.

```python


# Definir las variables numéricas y categóricas
numerical_vars = ['Budget', 'Spend', 'Impressions', 'Clicks', 'CTR', 'Conversions', 'CR', 'Revenue', 'ROI']
categorical_vars = ['Channel', 'Region', 'Product_Category', 'Market_Segment']

# Filtrar el DataFrame sin valores NaN en las variables categóricas
df_filtered = df.dropna(subset=categorical_vars)

# Imputar valores nulos en las variables numéricas con la mediana por grupo
for cat1 in df_filtered['Channel'].unique():  # Iterar sobre cada valor único de 'Channel'
    for cat2 in df_filtered['Region'].unique():  # Iterar sobre cada valor único de 'Region'
        for cat3 in df_filtered['Product_Category'].unique():  # Iterar sobre cada valor único de 'Product_Category'
            for cat4 in df_filtered['Market_Segment'].unique():  # Iterar sobre cada valor único de 'Market_Segment'
                # Filtrar los datos por grupo específico basado en las categorías actuales
                group = df_filtered[(df_filtered['Channel'] == cat1) &
                                    (df_filtered['Region'] == cat2) &
                                    (df_filtered['Product_Category'] == cat3) &
                                    (df_filtered['Market_Segment'] == cat4)]
                
                if not group.empty:  # Verificar si el grupo tiene datos
                    # Imputar la mediana de cada grupo a los valores nulos en las variables numéricas
                    for num_var in numerical_vars:
                        median_value = group[num_var].median(skipna=True)  # Calcular la mediana del grupo actual excluyendo NaNs
                        if pd.notnull(median_value):  # Verificar que la mediana calculada no sea NaN (mediana válida)
                            # Rellenar los valores nulos en el DataFrame original con la mediana calculada
                            df.loc[(df['Channel'] == cat1) &
                                   (df['Region'] == cat2) &
                                   (df['Product_Category'] == cat3) &
                                   (df['Market_Segment'] == cat4) &
                                   (df[num_var].isnull()), num_var] = median_value

# Imputar valores nulos restantes con la mediana general de cada variable numérica
for num_var in numerical_vars:
    if df[num_var].isnull().sum() > 0:  # Verificar si aún hay nulos en la variable numérica
        global_median = df[num_var].median(skipna=True)  # Calcular la mediana global de la variable numérica excluyendo NaNs
        df[num_var].fillna(global_median, inplace=True)  # Imputar los nulos con la mediana global

# Verificar si aún quedan nulos en las variables numéricas e imprimir el resultado
print(df[numerical_vars].isnull().sum())


```

### **5.2 - IMPUTACIÓN DE LAS VARIABLES CATEGÓRICAS CON RANDOM FOREST**

```python
# Importar las librerías necesarias
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder


# Paso 1: Imputar las variables numéricas utilizando la mediana

# Lista de variables numéricas
num_vars = ['Budget', 'Spend', 'Impressions', 'Clicks', 'CTR', 'Conversions', 'CR',
            'Revenue', 'ROI']

# Imputar valores nulos con la mediana
for col in num_vars:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

# Paso 2: Preparar los datos para la imputación de variables categóricas

# Lista de variables categóricas
cat_vars = ['Channel', 'Product_Category', 'Region', 'Market_Segment']

# Identificar variables categóricas con y sin valores nulos
cat_vars_with_nulls = [var for var in cat_vars if df[var].isnull().any()]
complete_cat_vars = [var for var in cat_vars if df[var].notnull().all()]

# Paso 3: Codificar las variables categóricas sin valores nulos

# Crear una instancia del codificador OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Ajustar y transformar las variables categóricas sin valores nulos
encoded_cat = encoder.fit_transform(df[complete_cat_vars])

# Convertir el resultado a un DataFrame y asignar los nombres de columnas
encoded_cat_df = pd.DataFrame(encoded_cat, columns=encoder.get_feature_names_out(complete_cat_vars), index=df.index)

# Paso 4: Combinar las variables numéricas y categóricas codificadas para crear las variables predictoras

# Combinar las variables numéricas y categóricas codificadas
X = pd.concat([df[num_vars], encoded_cat_df], axis=1)

# Asegurar que no haya valores nulos en las variables predictoras
X.fillna(0, inplace=True)  # Puedes ajustar este paso según tus necesidades

# Paso 5: Imputar cada variable categórica con valores nulos utilizando Random Forest

for var in cat_vars_with_nulls:
    # Separar los datos en entrenamiento y predicción
    df_notnull = df[df[var].notnull()].copy()
    df_null = df[df[var].isnull()].copy()
    
    # Preparar las variables predictoras y objetivo para el conjunto de entrenamiento
    X_train = X.loc[df_notnull.index]
    y_train = df_notnull[var]
    
    # Preparar las variables predictoras para el conjunto de predicción
    X_pred = X.loc[df_null.index]
    
    # Crear y entrenar el modelo Random Forest
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    
    # Predecir los valores faltantes
    y_pred = model.predict(X_pred)
    
    # Asignar los valores predichos al DataFrame original
    df.loc[df_null.index, var] = y_pred

# Paso 6: Verificar que no queden valores nulos en las variables categóricas

print("Valores nulos en variables categóricas después de la imputación:")
print(df[cat_vars].isnull().sum())

# Ahora, todas las variables numéricas han sido imputadas con la mediana,
# y las variables categóricas han sido imputadas utilizando Random Forest.

```

Imputar nulos en Campaign_Effectiveness

```python

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import root_mean_squared_error

# Copiar el DataFrame original para trabajar con él
df2 = df.copy()
scaler = StandardScaler()

# Escalar Campaign_Effectiveness
scaled_campaign_effectiveness = scaler.fit_transform(df2[['Campaign_Effectiveness']])
df2['Campaign_Effectiveness'] = scaled_campaign_effectiveness

# Variables numéricas a usar para reparar los nulos
columnas = ['Campaign_Effectiveness', 'Budget', 'Spend', 'Impressions', 'Clicks', 'CTR', 'Conversions', 'CR', 'Revenue', 'ROI']

# Codificar las variables categóricas
encoder = OneHotEncoder()
df_encoded = pd.DataFrame(
    encoder.fit_transform(df2[['Channel', 'Region', 'Product_Category', 'Market_Segment']]).toarray(),
    columns=encoder.get_feature_names_out(['Channel', 'Region', 'Product_Category', 'Market_Segment']),
    index=df2.index
)

# Concatenar las variables numéricas y las variables categóricas codificadas para el modelo
model_df = pd.concat([df2[columnas], df_encoded], axis=1)

# Separar los datos con y sin Campaign_Effectiveness
df_with_campaign_effectiveness = model_df.dropna(subset=['Campaign_Effectiveness'])
df_without_campaign_effectiveness = model_df[model_df['Campaign_Effectiveness'].isnull()].drop(columns='Campaign_Effectiveness')

# Dividir los datos de entrenamiento
X_train = df_with_campaign_effectiveness.drop(columns='Campaign_Effectiveness')
y_train = df_with_campaign_effectiveness['Campaign_Effectiveness']
X_test = df_without_campaign_effectiveness

# Encontrar el mejor valor de k usando validación cruzada
mse = []
K_range = range(1, 8)
for k in K_range:
    knn = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
    mse.append(scores.mean())
    print(f'k={k}, scores={scores}')

# Graficar los valores
plt.plot(K_range, mse)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated MSE')
plt.show()

# Seleccionar el mejor valor de k
best_k = K_range[mse.index(min(mse))]
print(f'El mejor valor de k es {best_k}')

# Imputar los valores nulos con el mejor modelo
knn = KNeighborsRegressor(n_neighbors=best_k)
knn.fit(X_train, y_train)
imputed_campaign_effectiveness = knn.predict(X_test)

# Asignar los valores imputados al DataFrame original
df.loc[df['Campaign_Effectiveness'].isnull(), 'Campaign_Effectiveness'] = scaler.inverse_transform(imputed_campaign_effectiveness.reshape(-1, 1))

# Verificar si aún quedan nulos en la columna imputada
print(df['Campaign_Effectiveness'].isnull().sum()/df.shape[0]*100)
```

Imputar nulos en Customer_Satisfaction

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt

# Copiar el DataFrame original para trabajar con él
df2 = df.copy()
scaler = StandardScaler()

# Escalar Customer_Satisfaction
scaled_customer_satisfaction = scaler.fit_transform(df2[['Customer_Satisfaction']])
df2['Customer_Satisfaction'] = scaled_customer_satisfaction

# Variables numéricas a usar para reparar los nulos
columnas = ['Customer_Satisfaction', 'Budget', 'Spend', 'Impressions', 'Clicks', 'CTR', 'Conversions', 'CR', 'Revenue', 'ROI']

# Codificar las variables categóricas
encoder = OneHotEncoder()
df_encoded = pd.DataFrame(
    encoder.fit_transform(df2[['Channel', 'Region', 'Product_Category', 'Market_Segment']]).toarray(),
    columns=encoder.get_feature_names_out(['Channel', 'Region', 'Product_Category', 'Market_Segment']),
    index=df2.index
)

# Concatenar las variables numéricas y las variables categóricas codificadas para el modelo
model_df = pd.concat([df2[columnas], df_encoded], axis=1)

# Imputar Customer_Satisfaction
# Separar los datos con y sin Customer_Satisfaction
df_with_customer_satisfaction = model_df.dropna(subset=['Customer_Satisfaction'])
df_without_customer_satisfaction = model_df[model_df['Customer_Satisfaction'].isnull()].drop(columns='Customer_Satisfaction')

# Dividir los datos de entrenamiento para Customer_Satisfaction
X_train_cs = df_with_customer_satisfaction.drop(columns='Customer_Satisfaction').dropna()
y_train_cs = df_with_customer_satisfaction.loc[X_train_cs.index, 'Customer_Satisfaction']
X_test_cs = df_without_customer_satisfaction

# Encontrar el mejor valor de k usando validación cruzada para Customer_Satisfaction
mse_cs = []
K_range = range(1, 8)
for k in K_range:
    knn = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn, X_train_cs, y_train_cs, cv=5, scoring='neg_mean_squared_error')
    mse_cs.append(scores.mean())
    print(f'k={k} (Customer_Satisfaction), scores={scores}')

# Seleccionar el mejor valor de k para Customer_Satisfaction
best_k_cs = K_range[mse_cs.index(min(mse_cs))]
print(f'El mejor valor de k para Customer_Satisfaction es {best_k_cs}')

# Imputar los valores nulos de Customer_Satisfaction con el mejor modelo
knn_cs = KNeighborsRegressor(n_neighbors=best_k_cs)
knn_cs.fit(X_train_cs, y_train_cs)
imputed_customer_satisfaction = knn_cs.predict(X_test_cs)

# Asignar los valores imputados al DataFrame original
df.loc[df['Customer_Satisfaction'].isnull(), 'Customer_Satisfaction'] = scaler.inverse_transform(imputed_customer_satisfaction.reshape(-1, 1))

# Verificar si aún quedan nulos en la columna imputada
print(df['Customer_Satisfaction'].isnull().sum()/df.shape[0]*100)
```
